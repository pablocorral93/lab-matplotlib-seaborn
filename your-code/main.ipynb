#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.arange(0,100)
y = x**2
y1 = np.exp(x)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot (x, y) on the first subplot
ax1.plot(x, y, color='blue', linewidth=2, linestyle='-', label='y = x^2')
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Plot 1',fontsize=16)
ax1.legend()

# Plot (x, y) on the second subplot
ax2.plot(x, y1, color='red', linewidth=2, linestyle='--', label='y = e^x')
ax2.set_yscale('log')
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('Plot 2',fontsize=16)
ax2.legend()

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

median_by_month = fitbit.groupby('Months_encoded')['Steps'].median()

median_by_month.plot()
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.title('Median Steps by month')
plt.show()


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

median_work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].median()

median_work_or_weekend.plot(kind='bar', color=['b', 'g']) 

plt.xlabel('Day Type')
plt.xticks(rotation ='horizontal') 
plt.ylabel('Median Steps')
plt.title('Median Steps on Workdays vs. Weekends')
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

feature_pairs = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

# Create a loop to plot each feature pair
for feature_x, feature_y in feature_pairs:
    plt.figure()  # Create a new figure for each plot
    plt.scatter(fitbit[feature_x], fitbit[feature_y])
    plt.xlabel(feature_x)
    plt.ylabel(feature_y)
    plt.title(f'{feature_x} vs {feature_y}')


plt.show()
```

```python
# your code here

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
only_numerics = titanic.select_dtypes(include=numerics)

correlation_matrix = only_numerics.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.show()
```
